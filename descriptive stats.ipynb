```python
def mean(col):
    sum = 0
    for i in col:
        sum += i
    return sum/len(col);
```

```python
mean(iris.SepalLengthCm)
```

```python
def median(column):
    sorted_col = np.sort(column,axis=0)
    val = sorted_col[math.floor(len(sorted_col)/2)] + sorted_col[math.ceil(len(sorted_col)/2)]
    return val/2
```

```python
median(iris.SepalLengthCm)
```

```python
def variance(column):
    avg = mean(column)
    x = 0
    for i in column:
        a = (avg - i)
        x += pow(a,2)
    return x/len(column)
```

```python
variance(iris.SepalLengthCm)
```

```python
def sd(column) :
    return pow(variance(column), 0.5)
```

```python
sd(iris.SepalLengthCm)
```

```python
def Cov(column):
    return sd(column)/mean(column)
```

```python
Cov(iris.SepalLengthCm)
```
